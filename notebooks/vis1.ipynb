```python
import pandas as pd
import plotly.graph_objects as go
```

```python
# datasets importen
df = pd.read_csv("../resources/us_accidents.csv")
accidents_per_week = pd.read_csv('../resources/Weekly_Accident_Counts.csv')
humidity = pd.read_csv('../resources/dataset_weather/us_humidity.csv')
temperature = pd.read_csv('../resources/dataset_weather/us_temperature.csv')
wind_speed = pd.read_csv('../resources/dataset_weather/us_wind_speed.csv')
mean_humid = pd.read_csv('../resources/dataset_weather/mean_humidity.csv')
mean_data = pd.read_csv('../resources/dataset_weather/mean_data.csv')
```

```python
def mean_calc(df):
    df['mean'] = df.iloc[:, 1:45253].mean(axis=1)
```

```python
mean_calc(humidity)
mean_calc(temperature)
mean_calc(wind_speed)
```

```python
mean_data_good = pd.DataFrame({'date':humidity['datetime'],'humidity':humidity['mean'],'temperature':temperature['mean'],'wind_speed':wind_speed['mean']})
```

```python
mean_data_good.to_csv('mean_data_good.csv', index=False)
```

```python
# Create the trace for the plot
trace1 = go.Scatter(x=accidents_per_week['Start_Time'], y=accidents_per_week['Accidents'], mode='lines', name='Accidents per Week')
trace2 = go.Scatter(x=mean_humid['date'], y=mean_humid['mean_humidity'], mode='lines', name='Accidents per Week')
# trace3 = go.Scatter(x=mean_data['date'], y=mean_data['mean_temperature'], mode='lines', name='Accidents per Week')
# trace4 = go.Scatter(x=mean_data['date'], y=mean_data['mean_wind_speed'], mode='lines', name='Accidents per Week')


# Create the figure and add the trace
fig = go.Figure(data=[trace1])
fig.add_scatter(x=mean_humid['date'], y=mean_humid['normalized_mean_humidity'])
# fig.add_scatter(x=mean_data['date'], y=mean_data['mean_temperature'])
# fig.add_scatter(x=mean_data['date'], y=mean_data['mean_wind_speed'])

# Update layout for better visualization
fig.update_layout(
    title='Weekly Number of Accidents',
    xaxis_title='Date',
    yaxis_title='Number of Accidents',
    xaxis=dict(tickformat='%Y-%m-%d'),
    template='plotly_white'
)

# Show the plot
fig.show()
```
